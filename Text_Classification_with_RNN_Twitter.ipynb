# Loading data and prepare data for modeling

```
#Load the data saved from drive


import os

imdb_dir = '/content/drive/Tweets Trump vs. Obama/'

labels = []
texts = []
count = 0
for label_type in ['Trump/', 'Obama/']:
    dir_name = os.path.join(imdb_dir, label_type)
    for fname in os.listdir(dir_name):
        if fname[-4:] == '.txt':
            count = count + 1
            print(count)
            f = open(os.path.join(dir_name, fname))
            texts.append(f.read())
            f.close()
            if label_type == 'Trump/':
                labels.append(0)
            else:
                labels.append(1)
```

```
# Tokenize, pad and prepare training and validation data
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
import numpy as np

maxlen = 100  # cut tweets after 100 words
training_samples = 2560  # training set includes 200 samples
validation_samples = 640  # validation set includes 10000 samples
max_words = 10000  #only consider the top 10,000 words in the dataset  


# Vectorize a text corpus by turning each text into either a sequence of integers
# omits common characters 
tokenizer = Tokenizer(num_words=max_words)

# fit_on_texts to learn the indices of the words
tokenizer.fit_on_texts(texts)

# the sequences are comprised of those indices
sequences = tokenizer.texts_to_sequences(texts)

word_index = tokenizer.word_index
print('Found %s unique tokens.' % len(word_index))

# pad sequences to gain uniform length in case of low word counts
data = pad_sequences(sequences, maxlen=maxlen)

# converted labels from list to a numpy array for modeling
labels = np.asarray(labels)
print('Shape of data tensor:', data.shape)
print('Shape of label tensor:', labels.shape)

# ensure that all sequences in a list have the same length
```

```
# Shuffle the data, split the data into a training set and a validation set
indices = np.arange(data.shape[0])
np.random.shuffle(indices)
data = data[indices]
labels = labels[indices]

x_train = data[:training_samples] # from 0 to (2560) training samples
y_train = labels[:training_samples]
x_val = data[training_samples: training_samples + validation_samples]
y_val = labels[training_samples: training_samples + validation_samples]
```

```
# save list into file for easily read

np.ndarray.tofile(data,"data1.npy")
np.ndarray.tofile(labels,"labels1.npy")
```

```
x_test = data[training_samples + validation_samples:]
y_test = labels[training_samples + validation_samples:]
```

# Modeling: Fit three models with different architecture

```
# Model1. LSTM + GRU

from keras.models import Sequential
from keras.layers import Embedding, Flatten, Dense
import tensorflow as tf
from tensorflow.keras.callbacks import EarlyStopping

embedding_dim = 8
#input_shape=(max_words, maxlen*embedding_dim),
model1 = tf.keras.Sequential()
model1.add(tf.keras.layers.Embedding(max_words, embedding_dim, input_length=maxlen))
model1.add(tf.keras.layers.LSTM(80, activation='relu',return_sequences=True))
model1.add(tf.keras.layers.GRU(40))
model1.add(tf.keras.layers.Flatten())
model1.add(tf.keras.layers.Dense(32, activation='relu'))
model1.add(tf.keras.layers.Dense(1, activation='sigmoid'))
model1.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['acc'])
model1.summary()


es = EarlyStopping(monitor='val_loss', mode='min',
                   patience=10, 
                   verbose=1)

history = model1.fit(x_train, y_train,
                    epochs=30,
                    batch_size=20,
                    validation_data=(x_val, y_val),
                    verbose=1,
                    callbacks=[es])
```

```
from sklearn.metrics import classification_report

y_pred_1 = model1.predict(x_test)

predict_1 = []
for i in y_pred_1:
  if i >= np.mean(y_pred_1):
    predict_1.append(1)
  else:
    predict_1.append(0)

print ("Mean value is: ", np.mean(y_pred_1),";", "Median value is: ", np.median(y_pred_1), "Predict size is:", y_pred_1.shape)
print ("Classify by Mean value", np.mean(y_pred_1),":")
print(classification_report(y_test, predict_1))


predict_1_2 = []
for i in y_pred_1:
  if i >= np.median(y_pred_1):
    predict_1_2.append(1)
  else:
    predict_1_2.append(0)
print ("Classify by Median value", np.median(y_pred_1),":")
print(classification_report(y_test, predict_1_2))
```

```

# Model2. LSTM 

embedding_dim = 8
model2 = tf.keras.Sequential()
model2.add(tf.keras.layers.Embedding(max_words,embedding_dim, input_length=maxlen))
model2.add(tf.keras.layers.LSTM(80, activation='relu',return_sequences=True))
model2.add(tf.keras.layers.Flatten())
model2.add(tf.keras.layers.Dense(32, activation='relu'))
model2.add(tf.keras.layers.Dense(1, activation='sigmoid'))
model2.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['acc'])
model2.summary()


es = EarlyStopping(monitor='val_loss', mode='min',
                   patience=10, 
                   verbose=1)

history2 = model2.fit(x_train, y_train,
                    epochs=30,
                    batch_size=20,
                    validation_data=(x_val, y_val),
                    verbose=1,
                    callbacks=[es])
```

```
y_pred_2 = model2.predict(x_test)

predict_2 = []
for i in y_pred_2:
  if i >= np.mean(y_pred_2):
    predict_2.append(1)
  else:
    predict_2.append(0)

print ("Mean value is: ", np.mean(y_pred_2),";", "Median value is: ", np.median(y_pred_2), "Predict size is:", y_pred_2.shape)
print ("Classify by Mean value", np.mean(y_pred_2),":")
print(classification_report(y_test, predict_2))


predict_2_2 = []
for i in y_pred_2:
  if i >= np.median(y_pred_2):
    predict_2_2.append(1)
  else:
    predict_2_2.append(0)
print ("Classify by Median value", np.median(y_pred_2),":")
print(classification_report(y_test, predict_2_2))
```

```
# Model3. Conv1D + LSTM

model3 = tf.keras.Sequential()

model3.add(tf.keras.layers.Embedding(max_words, embedding_dim, input_length=maxlen))

model3.add(tf.keras.layers.Conv1D(filters=80, kernel_size=3,activation='relu')) 
model3.add(tf.keras.layers.MaxPooling1D(pool_size=3))
model3.add(tf.keras.layers.Dropout(0.5))

model3.add(tf.keras.layers.LSTM(50, activation='relu'))
model3.add(tf.keras.layers.Dropout(0.5))

model3.add(tf.keras.layers.Flatten())
model3.add(tf.keras.layers.Dense(64, activation='relu'))
model3.add(tf.keras.layers.Dense(32, activation='relu'))
model3.add(tf.keras.layers.Dropout(0.5))
model3.add(tf.keras.layers.Dense(1, activation='sigmoid'))
model3.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['acc'])
model3.summary()


es = EarlyStopping(monitor='val_loss', mode='min',
                   patience=10, 
                   verbose=1)

history3 = model3.fit(x_train, y_train,
                    epochs=30,
                    batch_size=20,
                    validation_data=(x_val, y_val),
                    verbose=1,
                    callbacks=[es])

```

```
y_pred_3 = model3.predict(x_test)

predict_3 = []
for i in y_pred_3:
  if i >= np.mean(y_pred_3):
    predict_3.append(1)
  else:
    predict_3.append(0)

print ("Mean value is: ", np.mean(y_pred_3),";", "Median value is: ", np.median(y_pred_3), "Predict size is:", y_pred_3.shape)
print ("Classify by Mean value", np.mean(y_pred_3),":")
print(classification_report(y_test, predict_3))


predict_3_2 = []
for i in y_pred_3:
  if i >= np.median(y_pred_3):
    predict_3_2.append(1)
  else:
    predict_3_2.append(0)
print ("Classify by Median value", np.median(y_pred_3),":")
print(classification_report(y_test, predict_3_2))
```

# Analysis: Compare the results

1. The 'support' index shows the number of occurrence of 2 classes are: 412 and 388, which is a comparatively balanced data set.

2. According to precision score, 1st modeel correcty clarified 52% of Trump's tweets if setting cut-off value based on mean value of prediction values. While correctly predict null of Obama's tweets. So I additionally set median value as cut-off. The result shows more about this model: either for Trump or Obama, 1st model could only predict less than 50% of their tweets.

3. Then I tried to simplify my model by taking off the GRU layers. The performance of 2nd model is improved. Either meidan cut-off or mean cut-off shows higher than 93% accuracy.

4. For cut-off values, by comparing 3 models' results, I prefer median value as cut-off since median won't be affected by sample size.

5. When it comes to the 3rd models, since 2nd model proves LSTM has better performance, I add a convolutional layer and dropout function on top of the LSTM layer. While even the accuracy on predicting the class0 was improved to 96% (for 2nd model it's 95%), accuracy of class 1 dropped to 87% compare to 93% in 2nd model. So 2nd will be the best among these three models.
